# Cálculo de las métricas básicas de nodos

## Creamos los grafos

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.lollipop_graph(6, 4)

nx.draw(G, with_labels=True)
plt.show()
```

```python
# Uno dirigido

D = nx.gn_graph(10, seed = 0)
nx.draw(D, with_labels=True)
plt.show()
```

## Grado

```python
# Undirected

print("Grado de 8: %d" % nx.degree(G,8))
print("Grado de 2: %d" % nx.degree(G,2))
print("Grado de 5: %s" % nx.degree(G,5))
```

```python
# Directed. Observad que cambia la manera de hacer la llamada a la función.

print("InDegree de 1: %d" % D.in_degree(1))
print("OutDegree de 1: %d" % D.out_degree(1))
print("Degree de 1: %s" % D.degree(1))
```

```python
# Histograma de grados (van ordenados en grado 0, 1, 2, 3, ...)

print(nx.degree_histogram(G))
print(nx.degree_histogram(D))
```

```python
# Vecinos (adyacentes) al nodo 5
list(nx.neighbors(G,5))
```

```python
# En un dirigido, los vecinos son solo de salida
list(nx.neighbors(D,1))
```

```python
# Y así se obtienen todos los "colindantes"
list(nx.all_neighbors(D,1))
```
